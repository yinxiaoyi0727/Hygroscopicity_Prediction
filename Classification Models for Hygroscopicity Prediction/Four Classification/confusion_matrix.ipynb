```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# 读取结果文件
df = pd.read_csv("test_result_TabPFN_four_classification.csv")

# 获取真实标签和预测标签
y_true = df['true_label'].astype(str)
y_pred = df['predicted_label'].astype(str)

# 设置固定顺序
labels = ["Non-hygroscopic", "Slightly hygroscopic", "Hygroscopic", "Very hygroscopic"]
# labels = ["Non-hygro.", "Slightly-hygro.", "Hygro.", "Very-hygro."]

# 计算混淆矩阵
cm = confusion_matrix(y_true, y_pred, labels=[
    "no hygroscopic", "slightly hygroscopic", "hygroscopic", "very hygroscopic"
])

# 绘制热图并保存
plt.figure(figsize=(6, 5), dpi=600)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=labels, yticklabels=labels)

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.tight_layout()

# 保存为 PNG 文件
plt.savefig("4_class_confusion_matrix.png", dpi=600, bbox_inches="tight")

plt.show()
```
